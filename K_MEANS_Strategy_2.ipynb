```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# load data!
data = np.load('AllSamples.npy')

# set initial points!
k1= None
i_point1=None
k2=None
i_point2=None 

print(k1)
print(i_point1)
print(k2)
print(i_point2)
```

```python
# calculate the distance between two points
def distance (a,b):
    #return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)
    return math.sqrt(((a-b)**2).sum())

# cluster is a numpy array
def wcss(cluster):
    wcss=0.0
    mean = cluster.mean(axis=0)
    for point in cluster:
        wcss+= distance(point, mean)**2
    return wcss


def kmeans (data, k):
    best_obj_func = []
    
    best_centers = []         
    
    
    for j in range(10):        
                         
        centers = []


        #init = [[6.6384501, 8.33574252]]

        #init = [[3.79752017, 0.69134312]]   

        #"""
        init = []
        index= round(np.random.random()*(len(data)-1))

        init.append(data[index])
        #"""

        while len(init) < k:

            points = []  
            av_dists = []

            for point in data:

                if (init==point).any():
                    continue

                else:
                    points.append(point)

                    dist = 0           

                    l=0            

                    while l < len(init):                
                        dist += distance(point,init[l])
                        l+=1

                    av_dists.append(dist/len(init))

            index = av_dists.index(np.array(av_dists).max())

            init.append(points[index]) 

        centers.append(init) 

        iter = 0 

        while iter < 300:                

            clusters = []

            for i in range(k):

                clusters.append([])


            for point in data:

                distances=[]             

                for i in range(k):

                    dist = distance(point,centers[iter][i])
                    distances.append(dist)                

                ind = distances.index(np.array(distances).min())

                clusters[ind].append(point)     


            wcss_iter = 0.0

            new_centers = []      

            for cluster in clusters:
                wcss_iter += wcss(np.array(cluster))
                mu = np.array(cluster).mean(axis=0)             
                new_centers.append(mu) 

            centers.append(new_centers)            

            if (len(centers)>2) and (np.array_equal(centers[iter], centers[iter-1])):
                best_obj_func.append(wcss_iter)
                best_centers.append(centers[-1])
                break    

            iter+=1        

    very_best_index = best_obj_func.index(np.array(best_obj_func).min())
    return (best_obj_func[very_best_index], best_centers[very_best_index])  
```

```python
kmeans(data, 2)
```

```python
kmeans(data, 3)
```

```python
kmeans(data, 4)
```

```python
# compare with sklearn
k_means = KMeans(4)
k_means.fit(data)
print(k_means.inertia_)
print(k_means.cluster_centers_)
```
